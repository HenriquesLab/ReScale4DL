```python
from tifffile import imread, imwrite
from matplotlib import pyplot as plt

from downsampling.preprocessing.upscaling import upscale_img, upscale_labels
```

```python
img = imread("../../data/Staph/elyra_1/44nilered.tif")
label = imread("../../data/Staph/elyra_1/44label.tif")
```

```python
magnification = 4
upscaled_labels = upscale_labels(label, magnification)
upscaled_img = upscale_img(img, magnification)
```

```python
plt.imshow(upscaled_img[0])
```

```python
plt.imshow(upscaled_labels[0])
```

```python
imwrite("44nilered_upscaled.tif", upscaled_img)
imwrite("44label_upscaled.tif", upscaled_labels)
imwrite("44binarylabel_upscaled.tif", upscaled_labels>0)
```
